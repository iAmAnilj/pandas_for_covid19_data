## let's start working on dates

```python
italy_covid_df.date
```

Currently data type of date is **object**. First we need to convert this data type into date using **to_datetime** method from pandas.

```python
italy_covid_df['date'] = pd.to_datetime(italy_covid_df.date)
```

```python
italy_covid_df.date
```

Let's add new colums year,month,day and weekday using **DatetimeIndex** class from pandas.

```python
italy_covid_df['year'] = pd.DatetimeIndex(italy_covid_df.date).year
italy_covid_df['month'] = pd.DatetimeIndex(italy_covid_df.date).month
italy_covid_df['day'] = pd.DatetimeIndex(italy_covid_df.date).day
italy_covid_df['weekday'] = pd.DatetimeIndex(italy_covid_df.date).weekday
```

```python
italy_covid_df
```

Let's check overall matrix for the **May** month.

```python
# take sbset of may month from data frame
may_covid_df = italy_covid_df[italy_covid_df.month == 5]

# only take subset where summation need to perform
may_covid_matrix_df = may_covid_df[['new_cases','new_deaths','new_tests']]

#calculate total matrix of may
may_covid_matrix = may_covid_matrix_df.sum()
```

```python
may_covid_matrix
```

# Merging data from multiple sources

For performing merge and join we will import **`locations.csv`** file.

```python
urlretrieve('https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv', 
            'locations.csv')
```

```python
locations_df = pd.read_csv('locations.csv')
locations_df
```

```python
locations_df[locations_df.location.str.lower() == 'italy']
```

```python
italy_covid_df.head()
```

```python
italy_covid_df['location'] = 'Italy'
```

```python
italy_covid_df.head()
```

## Mergind Italy covid data to locations data on the location column.

```python
merge_df = italy_covid_df.merge(locations_df, on='location')
merge_df.head()
```

```python
merge_df['cases_per_million'] = merge_df.new_cases * 1e6 / merge_df.population
merge_df['deaths_per_million'] = merge_df.new_deaths * 1e6 / merge_df.population
merge_df['tests_per_million'] = merge_df.new_tests * 1e6 / merge_df.population
```

```python
merge_df.shape
```

```python
italy_covid_df.shape
```

```python
merge_df.head()
```

```python
merge_df.sort_values('deaths_per_million',ascending=False).head()
```

```python
merge_df['deaths_per_million'] = merge_df['deaths_per_million'].round()
```

```python
merge_df.head()
```

```python
merge_df.sort_values('deaths_per_million', ascending=False).head()
```

# Writing final data back to file

For writing data back in file we will take only those columns which is needed.
* for writing data back in CSV file we use **to_read** method from pandas.

```python
result_df = merge_df[['date',
                     'new_cases',
                     'new_deaths',
                     'new_tests',
                     'location','population',
                     'life_expectancy','hospital_beds_per_thousand',
                      'gdp_per_capita','cases_per_million','deaths_per_million','tests_per_million']]
```

```python
result_df.head()
```

```python
result_df
```

```python
result_df.to_csv('Final_result.csv', index=None)
```
